```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import  MinMaxScaler  #插值(missing value)和归一化(scaling values)
from sklearn.impute import SimpleImputer
# ML模型  ensemble集成
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
#用于超参数调整 Hyperparameter tuning  cv交叉验证
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
train_features = pd.read_csv('data/training_features.csv')
test_features = pd.read_csv('data/testing_features.csv')
train_labels = pd.read_csv('data/training_labels.csv')
test_labels = pd.read_csv('data/testing_labels.csv')
```

```python
print('Training Feature Size: ', train_features.shape)
print('Testing Feature Size:  ', test_features.shape)
print('Training Labels Size:  ', train_labels.shape)
print('Testing Labels Size:   ', test_labels.shape)
```

```python
# 插值(缺失值)
imputer = SimpleImputer(strategy='median')  #定义训练插值工具
imputer.fit(train_features)  #初始化（训练）插值工具
#将插值工具应用到train_data和test data上(因为是根据train上来预测test，所以采用train上定义的插值调整test中的missing)
X = imputer.transform(train_features)  #转换为np.ndarray对象了
X_test = imputer.transform(test_features)
```

```python
np.sum(np.isnan(X))
```

```python
np.sum(np.isnan(X_test))
```

```python
# 确保所有value都是finite实值
print(np.where(~np.isfinite(X)))
print(np.where(~np.isfinite(X_test)))
X.shape
```

### 特征缩放

```python
# 对于不同单位的数值，需要进行归一化使得能够参与到一个“意境”的算法，这使得单位不会影响算法
# 对于线性回归和随机森林则不需要特征缩放，对于SVM\K-means等需要计算“距离”的算法则需要缩放
# 方法一：标准化：X-U/sigma
# 方法二：归一化：(x-min)/(max-min),缩放到0-1
scaler = MinMaxScaler(feature_range=(0,1))
scaler.fit(X)
X = scaler.transform(X)
X_test = scaler.transform(X_test)
```

```python
y = np.array(train_labels).reshape((-1,)) #2D->1D
y_test = np.array(test_labels).reshape((-1,))
y.shape
```

### 选择ML模型进行评估

```python
#线性回归
#SVM回归
#随机森林回归
#梯度提升回归
#KNN回归

def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))

# test上评估模型
def fit_and_evaluate(model):
    
    # Train the model
    model.fit(X, y)
    
    # Make predictions and evalute
    model_pred = model.predict(X_test)
    model_mae = mae(y_test, model_pred)
    
    # Return the performance metric
    return model_mae
```

```python
lr = LinearRegression()
lr_mae = fit_and_evaluate(lr)
print('Linear Regression在测试集上的性能: MAE = %0.4f' % lr_mae)
```

```python
svm = SVR(C = 1000, gamma = 0.1)
svm_mae = fit_and_evaluate(svm)

print('Support Vector Machine Regression在测试集上的性能: MAE = %0.4f' % svm_mae)
```

```python
random_forest = RandomForestRegressor(random_state=60)
random_forest_mae = fit_and_evaluate(random_forest)

print('Random Forest Regression在测试集上的性能: MAE = %0.4f' % random_forest_mae)
```

```python
gradient_boosted = GradientBoostingRegressor(random_state=60)
gradient_boosted_mae = fit_and_evaluate(gradient_boosted)

print('Gradient Boosted Regression在测试集上的性能: MAE = %0.4f' % gradient_boosted_mae)
```

```python
knn = KNeighborsRegressor(n_neighbors=10,algorithm='ball_tree')
knn_mae = fit_and_evaluate(knn)

print('K-Nearest Neighbors Regression在测试集上的性能: MAE = %0.4f' % knn_mae)
```

```python
# 发现RF和GB性能最好。
model_comparison = pd.DataFrame({'model': ['Linear Regression', 'Support Vector Machine',
                                           'Random Forest', 'Gradient Boosted',
                                            'K-Nearest Neighbors'],
                                 'mae': [lr_mae, svm_mae, random_forest_mae, 
                                         gradient_boosted_mae, knn_mae]})

# 横向条形图
model_comparison.sort_values('mae', ascending = False).plot(x = 'model', y = 'mae', kind = 'barh',
                                                           color = 'green', edgecolor = 'black',alpha = 0.7)

# Plot formatting
plt.ylabel('')
plt.xlabel('Mean Absolute Error')
# plt.xticks(size = 14)
plt.title('Model Comparison on Test MAE', size = 20)
```

```python
# 在默认超参数下，随机森林最好，梯度增强次之（差距不大）
# 对SVM而言，其性能受超参数影响极大，而上面的两个模型性能不太依赖于模型设置的超参数

# 9.1659<<25,说明机器学习是使用的，我们采用的模型都优于baseline
```

### 以GBT进行超参数调优来优化出最佳模型

#### 优化方法：1、网格搜索（穷举参数选项）    2、随机搜索（随机组合提供的一系列参数选项）    3、K折交叉验证；一般先一后二（先缩减范围后进行有限穷举）

```python
# 超参数：自己设计的参数   模型参数：模型自己学习到的参数，我们看不到：如梯度，权重等
# 超参数调优：达到过拟合和欠拟合的平衡
# 欠拟合处理：加深模型复杂性：深度、数量等
# 过拟合处理：降低模型复杂度、正则化、加入丢失层drop layer，来加深模型的泛化能力。可以用样本的均值和方差与理想多类样本集合的均值、方差来理解
```

#### K折交叉验证CV用于评估超参数性能(将每个K次迭代的平均误差作为最终的性能指标)   （ 在K折之后立即在所有training data上训练模型）。多次采用不同超参数组合，每个组合都尝试K折，记录最终性能后比较不同超参组合的性能

```python
loss = ['ls', 'lad', 'huber']  #备选损失函数
n_estimators = [100, 500, 900, 1100, 1500]   #树量
max_depth = [2, 3, 5, 10, 15]   #每个树最大深度
min_samples_leaf = [1, 2, 4, 6, 8]  #每个叶子的最小样本量
min_samples_split = [2, 4, 6, 10]  #进行节点分割的最少样本量
max_features = ['auto', 'sqrt', 'log2', None]

# 定义网格

hyper_grid = {'loss': loss,
              'n_estimators': n_estimators,
              'max_depth': max_depth,
              'min_samples_leaf': min_samples_leaf,
              'min_samples_split': min_samples_split,
              'max_features': max_features}
```

```python
# 随机搜索对象
random_search_model = GradientBoostingRegressor(random_state = 42)#固定使用的随机数生成器，以便我们每次运行都能得到相同的结果
random_search_cv = RandomizedSearchCV(estimator = random_search_model,
                                      param_distributions = hyper_grid,
                                      cv = 4,
                                      n_iter = 25,#尝试不同组合的数量
                                      scoring = 'neg_mean_absolute_error',
                                      n_jobs = -1,
                                      verbose = 1,
                                      return_train_score = True,
                                      random_state=42)
```

```python
# 利用随机搜索缩减了超参数的搜索范围，进一步使用网格搜索进行进一步的超参数调优
# 创建网格搜索对象
gridAboutTree = {'n_estimators': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800]}
model = GradientBoostingRegressor(loss='huber',max_depth=5,min_samples_leaf=8,min_samples_split=6,max_features=None,random_state=42)#这里根据随机搜索确定一些超参数
# 对树的数量进行网格搜索
grid_search = GridSearchCV(estimator=model,param_grid=gridAboutTree,cv=4,\
                          scoring='neg_mean_absolute_error',verbose=1,\
                          n_jobs=-1,return_train_score=True)
```

```python
grid_search.fit(X,y)
```

## 模型结果解释

```python
# ML和DNN上构建的大部分是黑匣子模型，可解释性很差，不知道这些模型是怎么做出解释的
#方法：1、特征  2、用于模型解释的模型无关解释器Lime  3、尝试检查集成中的单个决策树，观察其节点划分
```

```python
# 从训练后的树集合中提取特征
feature_Results = pd.DataFrame({"feature":list(train_features.columns),\
                               "importance":my_model.feature_importances_}).sort_values("importance",ascending=False)
```

```python
feature_Results
```

```python
feature_Results = feature_Results.reset_index(drop=True)  #重新按行设置默认索引
feature_Results
```

## 结论：能源使用强度EUI和电力强度最重要，其他的重要性则极低，说明可以不必保留这些特征就能创建几乎性能和我们相同的模型

```python
feature_Results.loc[:9, :].plot(x = 'feature',
                                y = 'importance', 
                                edgecolor = 'k',
                                kind='barh',
                                color = 'green');
plt.xlabel('Importance');
plt.ylabel('')
plt.title('Feature Importances from Random Forest', size = 10)
```

```python
# 根据特征重要性进行特征选择（选择原特征的部分子集）
# 查实运用其重要的一些特征来训练模型
most_important_features = feature_Results['feature'][:10]
# 特征子集
indices = [list(train_features.columns).index(x) for x in most_important_features]

X_reduced = X[:, indices]
X_test_reduced = X_test[:, indices]

print('10个重要training features shape: ', X_reduced.shape)
print('10个重要esting  features shape: ', X_test_reduced.shape)
```

```python
lr = LinearRegression()

# 对照组使用所有基础特征
lr.fit(X, y)
lr_full_pred = lr.predict(X_test)

# 实验组使用10个重要特征
lr.fit(X_reduced, y)
lr_reduced_pred = lr.predict(X_test_reduced)

# Display results
print('对照组：Linear Regression Results: MAE =    %0.4f.' % mae(y_test, lr_full_pred))
print('实验组：Linear Regression Results: MAE = %0.4f.' % mae(y_test, lr_reduced_pred))
```

```python
# 说明精简特征并没有提高lr的性能，反而冗余的特征还能为其提供额外的信息，改善其性能
```

```python
reflect_my_model  =my_model
reflect_my_model.fit(X_reduced,y)
reflect_my_model_pred = reflect_my_model.predict(X_test_reduced)
print('对照组Results: MAE = %0.4f' % mae(y_test, my_pred))
print('实验组Results: MAE = %0.4f' % mae(y_test, reflect_my_model_pred))
```

```python
# 结果同理，但是性能下降不明显。我们可以通过减少特征来构建更简约的模型，此时训练速度会更快，且更易解释
```

```python
# Lime(Locally Interpretable Model-agnostic Explanations)通过在局部使用线性模型近似预测周围的区域来展示ML模型的思维方式
# 为了方便解释，使用10最重要特征训练的模型（牺牲性能，换取可解释性）

abs_residuals = abs(reflect_my_model_pred - y_test)
    
# 提取预测最成功和最失败的一些情况
wrong = X_test_reduced[np.argmax(abs_residuals), :]
right = X_test_reduced[np.argmin(abs_residuals), :]
```

```python
import lime 
import lime.lime_tabular
explainer = lime.lime_tabular.LimeTabularExplainer(training_data = X_reduced, 
                                                   mode = 'regression',
                                                   training_labels = y,
                                                   feature_names = list(most_important_features))
```

```python
wrong
```

```python
print('Prediction: %0.4f' % reflect_my_model.predict(wrong.reshape(1, -1)))
print('Real Value: %0.4f' % y_test[np.argmax(abs_residuals)])

# 解释wrong示例
wrong_explan = explainer.explain_instance(data_row = wrong, 
                                       predict_fn = reflect_my_model.predict)

# 预测解释
wrong_explan.as_pyplot_figure();
plt.title('Explanation of Prediction')
plt.xlabel('Effect on Prediction')
```

```python
wrong_explan.show_in_notebook(show_predicted_value=True)
```

```python
# 预测最差情况下，Prediction: 17.4321，Real Value: 96
# Lime图中显示了每个特征对最终预测的共线大小
# 途中显示 Site EUI<=0.39显著降低了预测值，而用电量>0.15则提高了预测值

# 可以这样解释，我们的模型认为score得分比实际得分高很多，因为EUI很高。但是实际得分却很高，这就告诉模型这是个明显的错误了，
#我们的模型因为EUI升高，而低估了建筑的score
```

```python
# 通过相同过程预测模型正确实例
print('Prediction: %0.4f' % reflect_my_model.predict(right.reshape(1, -1)))
print('Actual Value: %0.4f' % y_test[np.argmin(abs_residuals)])

# Explanation for wrong prediction
right_explan = explainer.explain_instance(data_row = right, predict_fn = reflect_my_model.predict)
right_explan.as_pyplot_figure()
plt.title('Explanation of Prediction')
plt.xlabel('Effect on Prediction')
```

```python
right_explan.show_in_notebook(show_predicted_value=True)
```
