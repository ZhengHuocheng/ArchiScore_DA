```python
# 发现每列的count值不同，说明有缺失值
# 缺失值处理
# -----计算每列缺失值占比

def missing_values_table(df):
        # 数据帧中总的null数
        mis_val = df.isnull().sum()
        
        # 缺失值占比
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # 临时表列名rename
        mis_val_table_ren_columns = mis_val_table.rename(\
            columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # 按百分比排序
        mis_val_table_ren_columns = mis_val_table_ren_columns[\
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(#筛选出含missing的行
        '% of Total Values', ascending=False).round(2)#四舍五入，保留1位小数
        
        # 消息提示
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # missing表
        return mis_val_table_ren_columns
```

### 特征工程最费时间，但回报最高，通常比模型使用的算法和超参数重要

```python
# 对数值特征进行了了对数变换和开方变换，结果只有对数才有不太明显的影响，而开发影响更小
# 对类别特征进行了独热编码（borough & property use type）
# 对于共线的特征可以删除来降低耗时


features = data.copy()

numeric_subset = data.select_dtypes('number')

for col in numeric_subset.columns:
    if col == 'score':
        next
    else:
        numeric_subset['log_' + col] = np.log(numeric_subset[col])
        
categorical_subset = data[['Borough', 'Largest Property Use Type']]

categorical_subset = pd.get_dummies(categorical_subset)

features = pd.concat([numeric_subset, categorical_subset], axis = 1)

features.head(5)
```

```python
# 删除共线性特征（多重共线性特征间相关系数高）
# site EUI 和 Weather_ther norm EUI之间高度相关，r=0.8
plot_data = data[['Weather Normalized Site EUI (kBtu/ft²)', 'Site EUI (kBtu/ft²)']].dropna()
plt.plot(plot_data['Site EUI (kBtu/ft²)'], plot_data['Weather Normalized Site EUI (kBtu/ft²)'], 'go')
plt.xlabel('Site EUI')
plt.ylabel('Weather Norm EUI')
# test = np.corrcoef(data[['Weather Normalized Site EUI (kBtu/ft²)', 'Site EUI (kBtu/ft²)']].dropna())
# test[0][1]
plt.title('Weather Norm EUI--Site EUI, R = %0.4f' % np.corrcoef(data[['Weather Normalized Site EUI (kBtu/ft²)', 'Site EUI (kBtu/ft²)']].dropna())[0][1])
```

```python
# 这种高度共线性的变量的多余的，保存其一即可，有多种方法
# 将共线性阈值设置为0.6，超过则留其一

def remove_collinear_features(x, threshold):
    y = x['score']
    x = x.drop(columns = ['score'])
    
    # 计算correlation matrix
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # 通过相关矩阵进行过滤
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)
            
            # If correlation exceeds the threshold
            if val >= threshold:
                drop_cols.append(col.values[0])

    # 去除相关pairs
    drops = set(drop_cols)
    print(drops)
    x = x.drop(columns = drops)
    x = x.drop(columns = ['Weather Normalized Site EUI (kBtu/ft²)', 
                          'Water Use (All Water Sources) (kgal)',
                          'log_Water Use (All Water Sources) (kgal)',
                          'Largest Property Use Type - Gross Floor Area (ft²)'])
    x['score'] = y
               
    return x
```

```python
features = remove_collinear_features(features, 0.6)
```

```python
# 进一步去除含全是nan的列，进一步降维
features = features.dropna(axis=1,how='all')#axis指明按照行删除还是按照列删除
features.shape   #降了40维度左右
```

```python
# 独热编码只是使得数据能够参与欧氏距离的运算，而树模型的分类标准可以是类型也可以是数值，、
# 所以对树模型而言，类别编码没有什么意义

# PCA能根据线性进行降维，也用于特征选择，但是这个特征是映射得到，没办法解释
```

```python
# 训练集测试集划分
no_score = features[features['score'].isna()]
score = features[features['score'].notnull()]

print(no_score.shape)
print(score.shape)  #选择有score的实例参与训练
```

```python
# 自变量、因变量分离
features = score.drop(columns='score')
targets = pd.DataFrame(score['score'])
# inf缺失值插补。注意inf毕竟不是nan，只是其包含的信息无法显示出来，是有信息的，这里将其转换为nan，只是便于之后的插值
features = features.replace({np.inf: np.nan, -np.inf: np.nan})

# 数据计划分7：3
X, X_test, y, y_test = train_test_split(features, targets, test_size = 0.3, random_state = 42)

print(X.shape)
print(X_test.shape)
print(y.shape)
print(y_test.shape)
```

### 建立baseline

```python
# 针对回归问题，朴素的基线设置为训练集中的target的中位数。希望我们预测的指标高于这个中位数
# 中位数基线是简单方法，提供了较低的基线
# 建立指标平均绝对值误差MAE
def mae(y_true,y_pred):
    return np.mean(abs(y_true - y_pred))
```

```python
baseline_guess = np.mean(y)
baseline_guess
```

```python
# 计算测试集对baseline的平均估计偏差
org_mae = mae(y_test, baseline_guess)
org_mae
```

```python
# 说明测试集对baseline的平均估计偏差为25个点。
```
